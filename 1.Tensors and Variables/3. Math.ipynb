[Tensor flow math](https://www.tensorflow.org/api_docs/python/tf/math)

```python
import tensorflow as tf
```

**abs**

```python
x = tf.constant([-2.25, 3.25])
tf.abs(x)
```

```python
# complex number
x = tf.constant([[-2.25 + 4.75j], [-3.25 + 5.75j]])
tf.abs(x)
```

**Addition**

```python
x = [1, 2, 3, 4, 5]
y = 1
tf.add(x, y)
```

```python
x = tf.convert_to_tensor([1, 2, 3, 4, 5])
y = tf.convert_to_tensor(1)
x + y
```

```python
x = [1, 2, 3, 4, 5]
y = tf.constant([1, 2, 3, 4, 5])
tf.add(x, y)
```

**ceil**

```python
tf.math.ceil([-1.7, -1.5, -0.2, 0.2, 1.5, 1.7, 2.0])
```

**confusion_matrix**

```python
  tf.math.confusion_matrix([1, 2, 4], [2, 2, 4])
```

**divide**

```python
x = tf.constant([16, 12, 11])
y = tf.constant([4, 6, 2])
tf.divide(x,y)
```

**equal**

```python
x = tf.constant([2, 4])
y = tf.constant(2)
tf.math.equal(x, y)
```

```python
x = tf.constant([2, 4])
y = tf.constant([2, 4])
tf.math.equal(x, y)
```

**Maximum**

```python
x = tf.constant([0., 0., 0., 0.])
y = tf.constant([-2., 0., 2., 5.])
tf.math.maximum(x, y)
```

```python
x = tf.constant([-5., 0., 0., 0.])
y = tf.constant([-3.])
tf.math.maximum(x, y)
```

**minimum**

```python
x = tf.constant([0., 0., 0., 0.])
y = tf.constant([-5., -2., 0., 3.])
tf.math.minimum(x, y)
```

**pow**

```python
x = tf.constant([[2, 2], [3, 3]])
y = tf.constant([[8, 16], [2, 3]])
tf.pow(x, y)  # [[256, 65536], [9, 27]]
```

**real**

```python
x = tf.constant([-2.25 + 4.75j, 3.25 + 5.75j])
tf.math.real(x)  # [-2.25, 3.25]
```

**sqrt**

```python
x = tf.constant([[4.0], [16.0]])
tf.sqrt(x)
```

```python
y = tf.constant([[-4.0], [16.0]])
tf.sqrt(y)
```

```python
z = tf.constant([[-1.0], [16.0]], dtype=tf.complex128)
tf.sqrt(z)
```
