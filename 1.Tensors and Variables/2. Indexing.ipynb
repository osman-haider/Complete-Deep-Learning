```python
import tensorflow as tf
```

```python
index_tensor = tf.constant([2,3,5,1,7,0])
print(index_tensor)
print(index_tensor[1])
print(index_tensor[0:3])
```

```python
two_d_tensor = tf.constant([
    [1,2,3],
    [2,3,4],
    [3,4,5],
    [4,5,6]
])
print(two_d_tensor[0:4,0:2])
```

```python
print(two_d_tensor[0:3,:])
```

```python
print(two_d_tensor[...,0:2])
```
